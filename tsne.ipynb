```python
from sklearn import datasets
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
digits = datasets.load_digits()
X = digits.data[:500]
y = digits.target[:500]
```

```python
target_ids = range(len(digits.target_names))

for iters in [250, 1000, 2000]:
    tsne = TSNE(n_components=2,perplexity=30, random_state=0,  n_iter=iters)
    Reduced_2d = tsne.fit_transform(X)
    plt.figure(figsize=(6, 5))
    colors = 'r', 'g', 'b', 'c', 'm', 'y', 'k', 'w', 'orange', 'purple'
    for i, c, label in zip(target_ids, colors, digits.target_names):
        plt.scatter(Reduced_2d[y == i, 0], Reduced_2d[y == i, 1], c=c, label=label)
    print("Scatter plot for iteration =",iters)
    plt.legend()
    plt.show()
```

##### According to my observation having fixed perplexity=30 and varied number of iterations (100, 1000 and 2000) produces the same number of clusters , on scatter plot giving 9 cluster always although with very different shapes .The images above show three different runs at perplexity 30. The first one with  250 steps stopped before stability.  If we see a t-SNE plot with strange pinched shapes, chances are the process was stopped too early. Unfortunately, theres no fixed number of steps that yields a stable result. so we see different shapes
